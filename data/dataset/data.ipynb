```python
import pandas as pd

# Load the CSV file
file_path = 'ratings_restaurant.csv'
ratings_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
ratings_df.head()
```

```python
# Group by restaurant_id and find users who rated the same restaurant
grouped_restaurants = ratings_df.groupby('restaurant_id')['user_id'].agg(list)

# Filter out restaurants rated by more than one user
common_ratings = grouped_restaurants[grouped_restaurants.apply(lambda x: len(set(x)) > 1)]

# Display the result
common_ratings
```

```python
from collections import defaultdict

# Create a dictionary to store the set of restaurant_ids rated by each user
user_restaurant_dict = ratings_df.groupby('user_id')['restaurant_id'].apply(set).to_dict()

# Reverse the dictionary to find users with the same set of restaurant_ids
restaurants_users_dict = defaultdict(list)
for user, restaurants in user_restaurant_dict.items():
    restaurants_users_dict[frozenset(restaurants)].append(user)

# Find and display groups of users who rated the exact same set of restaurants
common_user_groups = {k: v for k, v in restaurants_users_dict.items() if len(v) > 1}
common_user_groups
```

```python
# Find users with unique sets of restaurant ratings
unique_user_groups = {k: v for k, v in restaurants_users_dict.items() if len(v) == 1}

# Display the unique user groups
unique_user_groups

```
