```python
import pandas as pd

# Membaca file CSV
file_path = 'Kuesioner Penelitian Sistem Rekomendasi Restaurant di Yogyakarta (Jawaban) - Form Responses 1.csv'
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama untuk melihat struktur data
df.head()
```

```python
# Melakukan unpivot pada data
df_melted = df.melt(id_vars=['user_id', 'Nama', 'Domisili'], var_name='Nama Restoran', value_name='rating')

# Mengganti nama kolom sesuai permintaan
df_melted.columns = ['user_id', 'nama', 'domisili', 'Nama Restoran', 'rating']

# Menampilkan hasil
df_melted.head()
```

```python
# Menambahkan user_id unik untuk setiap nama
df_melted['user_id'] = df_melted['nama'].astype('category').cat.codes + 1

# Mengurutkan data berdasarkan user_id dan nama restoran
df_melted = df_melted.sort_values(by=['Nama Restoran'])

# Menampilkan hasil
df_melted.head()

# df_melted.to_csv('ratings_restaurant.csv',index=False)
```
